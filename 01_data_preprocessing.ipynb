```python
import pandas as pd
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.info()
```

```python
movies.info()
```

```python
movies = movies[["id","genres","keywords","title","tagline","overview","production_companies"]]
credits = credits[["movie_id","cast","crew"]]
```

```python
movies.info()
```

```python
credits.info()
```

```python
# Lets join both dataframes
# how = 'inner' Keeps only matching rows from both DataFrames.If a row doesn’t match either from table it gets removed!!
df = pd.merge(movies, credits, left_on='id', right_on='movie_id', how='inner')
```

```python
df.info()
```

```python
df = df.drop(columns = ["movie_id"])
```

```python
df.info()
```

```python
desc = df.loc[0, 'genres']
```

```python
desc
```

```python
import ast
names = []
x = ast.literal_eval(desc)
# x = [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# item = {"id": 28, "name": "Action"}  when index = 0
for item in x:
    names.append(item['name'])

comma_desc = ','.join(names)
comma_desc
```

```python
import ast
def get_names(desc):
    names = []
    x = ast.literal_eval(desc)
    for item in x:
        names.append(item['name'])
    
    comma_desc = ','.join(names)
    return comma_desc
```

```python
df['genres'] = df['genres'].apply(get_names)
```

```python
df['genres']
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
```

```python
df['keywords'] = df['keywords'].apply(get_names)
df['cast'] = df['cast'].apply(get_names)
df['crew'] = df['crew'].apply(get_names)
df['production_companies'] = df['production_companies'].apply(get_names)
```

```python
df.sample()
```

```python
#need to replace space by '-' in production_companies, cast, crew because we want to treat them as single word while creating bag of words. 
df['genres'] = df['genres'].str.replace(' ', '-')
df['keywords'] = df['keywords'].str.replace(' ', '-')
df['production_companies'] = df['production_companies'].str.replace(' ', '-')
df['cast'] = df['cast'].str.replace(' ', '-')
df['crew'] = df['crew'].str.replace(' ', '-')
```

```python
df.sample()
```

```python
df.fillna(" ", inplace=True)

df['tags'] = df['genres'] + ',' + df['keywords'] + ',' + df['tagline'] + ',' + df['overview'] + ',' + df['production_companies'] + ',' + df['cast'] + ',' + df['crew']
```

```python
df.loc[0,'tags']
```

```python
df = df[["id","title","tags"]]
```

```python
df['tags'] = df['tags'].str.lower()
```

```python
df
```

```python
df['tags'] = df['tags'].str.replace('.', '') 
```
